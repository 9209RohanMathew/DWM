```python
q1 = w.quantile(0.25)
q3 = w.quantile(0.75)
print(q1)
print(q3)
```

```python
iqr = q3 - q1
lower_limit = q1-(1.5*iqr)
upper_limit = q3+(1.5*iqr)
print(iqr)
print(lower_limit)
print(upper_limit)
```
